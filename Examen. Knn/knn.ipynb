```python
import seaborn as sns
import matplotlib.pyplot as plt

def display_numeric():
    # Seleccionamos solo las columnas numéricas
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

    # Configuramos el tamaño de la figura
    plt.figure(figsize=(15, 10))

    # Iteramos sobre cada columna numérica y mostramos su distribución
    for i, col in enumerate(numeric_columns, 1):
        plt.subplot(2, 4, i)
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f'Distribución de {col}')
        plt.xlabel(col)
        plt.ylabel('Frecuencia')

    plt.tight_layout()
    plt.show()

display_numeric()
```

Pasamos sexo a binario

```python
# Convertir 'Sex'a 0s y 1s
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

Redondeamos los datos menores a 0 

```python
df['Age'] = df['Age'].round().astype(float)
```

Normalizamos con min / max

```python

df['Age'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())

df.head()
```

Normalizamos parch

```python
#Pasamos los que no son numericos a algo a partir de la mediana
df['Parch'] = pd.to_numeric(df['Parch'], errors='coerce')


parch_median = df['Parch'].median()
df['Parch'].fillna(parch_median, inplace=True)


df['Parch'] = df['Parch'].astype(int)


parch_distribution_after_imputation = df['Parch'].value_counts()
```

```python
#Ahora los datos atipicos, caso en que tienen parientes mayores a 9

df.loc[df['Parch'] > 9, 'Parch'] = 0


parch_distribution_adjusted_again = df['Parch'].value_counts()
df.describe()
```

Normalizamos Ticket 

Identificar valores  numéricos en la columna 'Ticket'

Codificar

```python
# Codificar el componente de cadena a un valor numérico (label)
from sklearn.calibration import LabelEncoder


df['Ticket_string'] = df['Ticket'].str.extract('([a-zA-Z]+)', expand=False).fillna('')
label_encoder = LabelEncoder()
df['Ticket_string_encoded'] = label_encoder.fit_transform(df['Ticket_string'])

# Combinar el label codificado con el componente numérico para obtener un nuevo valor de ticket en formato flotante
df['New_Ticket'] = df['Ticket_string_encoded'].astype(str) + df['Ticket'].str.extract('(\d+)', expand=False).fillna('0')
```

```python
df.drop(["Ticket_string","Ticket_string_encoded","Ticket"], axis=1, inplace=True)
df["Ticket"]=df["New_Ticket"]
df.drop(["New_Ticket"], axis=1, inplace=True)
df['Ticket'] = df['Ticket'].astype(float) #Pasamos a flotante
```

Intercambiar los datos que se encuentran entre Fare y Cabin

```python


# Identificar filas donde 'Cabin' es nulo y 'Fare' parece ser un código de cabina
condicion = df['Cabin'].isnull() & (df['Fare'].str.isnumeric() == False)

# Mover valores de 'Fare' a 'Cabin' para estas filas y establecer 'Fare' a NaN
df.loc[condicion, 'Cabin'] = df.loc[condicion, 'Fare']
df.loc[condicion, 'Fare'] = None

# Verificar la corrección mostrando las filas que fueron corregidas
correcciones = df[condicion]


# Identificar y reemplazar valores no numéricos en 'Fare' con NaN
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')

# Imputar valores NaN en 'Fare' con la mediana
fare_median = df['Fare'].median()
df['Fare'].fillna(fare_median, inplace=True)

print(correcciones[['Fare', 'Cabin']])
```

Imputar nulos de Fare

```python
median_fares = df.groupby('Pclass')['Fare'].median()

# Imputar los valores faltantes en 'Fare' usando la mediana correspondiente de 'Pclass'
for pclass, median_fare in median_fares.items():
    condition = (df['Pclass'] == pclass) & (df['Fare'].isnull())
    df.loc[condition, 'Fare'] = median_fare

# Verificar si aún existen valores faltantes en 'Fare'
missing_fares_after_imputation = df['Fare'].isnull().sum()
```

Cabin y Embarked intercambiados, al interambiar es necesario llenar nulos o valores que no correspondian a nada

```python
# Identificar filas donde 'Cabin' tiene valores "S", "C" o "Q"
filas_cabin_incorrectas = df['Cabin'].isin(['S', 'C', 'Q'])

# Intercambiar valores entre 'Cabin' y 'Embarked' para estas filas
for idx in df[filas_cabin_incorrectas].index:
    df.at[idx, 'Cabin'], df.at[idx, 'Embarked'] = df.at[idx, 'Embarked'], df.at[idx, 'Cabin']

# Imputar valores faltantes en 'Embarked' con la moda
moda_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(moda_embarked, inplace=True)

# Verificar la corrección mostrando las filas que fueron corregidas
filas_corregidas_cabin = df.loc[filas_cabin_incorrectas]
filas_corregidas_cabin[['Cabin', 'Embarked']]
```
